# Play w/ GPT-2 on OpenWebText w/ Distributionally Robust Optimization

```python
from huggingface_hub import login, get_full_repo_name

with open('huggingface.key') as f: hf_token = f.read().strip()
login(token=hf_token, write_permission=True, add_to_git_credential=True)

model_name = "gpt2-openwebtext-dro-test"
repo_name = get_full_repo_name(model_name)
repo_name
```

```python
import torch

from transformers import pipeline

device = torch.device("cuda:0") if torch.cuda.is_available() else torch.device("cpu")
pipe = pipeline(
    "text-generation", model=repo_name, device=device
)
```

```python
txt = 'One must fulfill the duties of squire, insofar as'
print(pipe(txt, num_return_sequences=1, max_new_tokens=128)[0]["generated_text"])
```
